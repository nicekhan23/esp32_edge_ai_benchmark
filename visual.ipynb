```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Visualizing a Window from TRIANGLE_ml.csv

```python
df_triangle = pd.read_csv('/content/TRIANGLE_ml.csv')
# Select the first window (row 0) and its sample data (columns from 'sample_0' onwards)
window_data_triangle = df_triangle.iloc[0, 2:]

plt.figure(figsize=(10, 4))
plt.plot(window_data_triangle.index, window_data_triangle.values)
plt.title('First Window from TRIANGLE_ml.csv (Sample Data)')
plt.xlabel('Sample Index (e.g., sample_0, sample_1, ...)')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
